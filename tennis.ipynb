```python
import pandas as pd
import numpy as np
from pathlib import Path
import torch
device = torch.device("cuda:0")
```

```python
winner_id = []
loser_id = []
surface = []

folder = Path("atp_matches")
for fil in folder.iterdir() :
    df=pd.read_csv(fil)
    winner_id.extend((df["winner_id"].tolist()))

    loser_id.extend((df["loser_id"].tolist()))

    surface.extend((df["surface"].tolist()))
winner_id_np=np.array(winner_id)
t_w = torch.tensor(winner_id_np,device=device)

loser_id_np=np.array(loser_id)
t_l = torch.tensor(loser_id_np,device=device)

surface_np=np.array(surface)
print(type(surface[0]))
surface_nr = []
for i in surface:
    if i == "Hard":
        surface_nr.append(0)
    elif i == "Clay":
        surface_nr.append(1)
    elif i == "Grass":
        surface_nr.append(2)
    elif i == "Carpet":
        surface_nr.append(3)
    else: surface_nr.append(4)
t_s = torch.tensor(surface_nr,device=device)


atp_ranks = []
df=pd.read_csv("atp_rankings_current.csv")
atp_ranks.extend((df["points"].tolist()))
atp_ranks_np=np.array(atp_ranks)
t_r = torch.tensor(atp_ranks_np)

atp_ranks_player = []
df=pd.read_csv("atp_rankings_current.csv")
atp_ranks_player.extend((df["player"].tolist()))
atp_ranks_player_np=np.array(atp_ranks_player)
t_r_p = torch.tensor(atp_ranks_player_np,device=device)




print(len(t_l),len(t_w),len(t_s))

```

```python
unik_players = []
for id in winner_id:
    allere_i = False
    for p in unik_players:
        if p == id:
            allere_i = True
            break
    if allere_i == False:
        unik_players.append(id)
    
for id in loser_id:
    allere_i = False
    for p in unik_players:
        if p == id:
            allere_i = True
            break
    if allere_i == False:
        unik_players.append(id)

id_til_embed_id = {}
c=0
for id in unik_players:
    id_til_embed_id[id] = c
    c = c + 1

winner_nr = []
loser_nr  = []

for pid in winner_id:
    winner_nr.append(id_til_embed_id[pid])

for pid in loser_id:
    loser_nr.append(id_til_embed_id[pid])

t_w_nr = torch.tensor(winner_nr, dtype = torch.long, device=device)
t_l_nr = torch.tensor(loser_nr,dtype=torch.long,device=device)
```
